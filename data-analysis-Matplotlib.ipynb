<h2>Matplotlib<h2>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Temel Çizimler

x eksenine göre y eksenini çizelim

```python
plt.plot([2,4,6,8], [3,5,7,9])
plt.show()
```

```python
x=[1,2,3]
y=[1,5,9]
plt.plot(x,y)
plt.show()
```

Başlık ekleme

```python
plt.title("Grafik")
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.plot(x,y)
plt.show()
```

x ve y ekseni etiketi belirleme

```python
plt.xticks([1,2,3])   
plt.yticks([1,5,9])
plt.title("Grafik")
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.plot(x,y)
plt.show()
```

<h4>Lejant Ekleme<h4>

```python
plt.plot(x,y, label='day')
plt.xticks([1,2,3])   
plt.yticks([1,5,9])
plt.title("Grafik")
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.plot(x,y)
plt.legend()
plt.show()
```

Çizgi rengi, kalınlığı, stili ve işaretini değiştirme

```python
plt.plot(x,y, label='day', color='red',linewidth=2, linestyle='--',marker='.')
plt.xticks([1,3,5])   
plt.yticks([1,5,10])
plt.title("Grafik")
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.plot(x,y)
plt.legend()
plt.show()
```

<h3>Tek grafikte 2 çizim yapma<h3>

```python
plt.plot(x,y, label='day', color='blue', linewidth = 2, linestyle = '--', marker='.')
plt.xticks([1,2,3,4,5])   
plt.yticks([1,4,9,16,25])
plt.title("İki Grafik Çizme")
plt.xlabel('x axis')
plt.ylabel('y ekseni')

x2 = np.arange(0,5,1)
plt.plot(x2,x2*2,color='green', linewidth = 2, marker='.', label='week')

plt.legend()
plt.show()
```

Grafiği Bilgisayara Kaydetme

```python
plt.plot(x,y, label='day', color='blue', linewidth = 2, linestyle = '--', marker='.')
plt.xticks([1,2,3,4,5])   
plt.yticks([1,4,9,16,25])
plt.title("İki Grafik Çizme")
plt.xlabel('x axis')
plt.ylabel('y ekseni')

x2 = np.arange(0,5,1)
plt.plot(x2,x2*2,color='green', linewidth = 2, marker='.', label='week')

plt.legend()
plt.savefig('MyGraph.png',dpi=250)
plt.show()
```

<h3>Barchart<h3>

```python
x = ['Denizli', 'Burdur', 'Isparta']
y = [100,27,45]

plt.bar(x,y)
plt.show()
```

```python
x = ['Denizli', 'Burdur', 'Isparta']
y = [100,27,45]
desen = plt.bar(x,y)
desen[0].set_hatch('/')
desen[1].set_hatch('|')
desen[2].set_hatch('.')                 
plt.show()
```
